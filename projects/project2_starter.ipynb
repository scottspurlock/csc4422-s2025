```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

## Read in the public and competition data

```python
public_url = 'https://drive.google.com/uc?export=download&id=1MkHSbPyKTCK_uVP269Uf_IM8ReGvpllG'
competition_url = 'https://drive.google.com/uc?export=download&id=1XoQ96X4jXejwSd6HAs4QoJwT2vBD8xq2'

df = pd.read_csv(public_url)
df_competition = pd.read_csv(competition_url)

df.head()
```

## Note that the competition data doesn't include the target, Beverage, since the goal of the competition is to predict this.

```python
df_competition.head()
```

## Next would be data wrangling
For simplicity, I'll just select a few numeric columns that are already ok for making a model. You'll want to do more to wrangle the data into shape.

```python
X = df[['Weight', 'ZipCode', 'Shoes']]
y = df['Beverage']

X.head()
```

## Make sure to do the same things to the competition data as you do to the public data

Except that there is no Beverage column in the test data, so we don't have a ycompetition

```python
Xcompetition = df_competition[['Weight', 'ZipCode', 'Shoes']]
Xcompetition.head()
```

## Next we would want to perform experiments with the public data to find the best type of model and best hyperparameters
We would want to split the training data into 2 sets to check how well things are working.

For this simple example, I'm just going to train a default KNN model with all the public data.

```python
model = KNeighborsClassifier()
model.fit(X, y)
```

## Once the final model is trained, we can predict the competition data

```python
ypred = model.predict(Xcompetition)
print('First 10 predictions:', ypred[:10])
```

## For submission to the Kaggle competition, we need to create a csv file
The format will have a column for ID and one for predicted Beverage, and one row for each test example.

### Notes
- Kaggle is expecting 500 predictions. Make sure not to drop any of the rows from the competition data.
- Kaggle is expecting predictions to be strings like 'Coffee'

```python
output = pd.DataFrame({'ID': df_competition['ID'], 'Beverage': ypred})

# Note that Kaggle is expecting the predictions to be strings like 'Coffee' and
# not numbers like 0. You may need to map output['Beverage'] to strings.
output
```

Save the predictions as a CSV file

```python
output.to_csv('submission.csv', index=False)
print("Kaggle submission was successfully saved...")
```
